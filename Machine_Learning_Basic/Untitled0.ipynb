```python
import netCDF4 as nc
import numpy as np

# Tạo một file NetCDF mới
file_path = '/content/drive/MyDrive/Data/VNP03MOD_NRT.A2023329.0512.002.2023329062910.nc'
dataset = nc.Dataset(file_path, 'w', format='NETCDF4')

# Tạo các chiều và biến
time_dim = dataset.createDimension('time', None)  # 'None' cho phép chiều có kích thước linh động
lat_dim = dataset.createDimension('latitude', 180)
lon_dim = dataset.createDimension('longitude', 360)

time_var = dataset.createVariable('time', 'f4', ('time',))
lat_var = dataset.createVariable('latitude', 'f4', ('latitude',))
lon_var = dataset.createVariable('longitude', 'f4', ('longitude',))

data_var = dataset.createVariable('temperature', 'f4', ('time', 'latitude', 'longitude'))

# Điền dữ liệu vào các biến
time_var[:] = np.arange(10)  # Ví dụ: 10 thời điểm
lat_var[:] = np.linspace(-90, 90, 180)  # Ví dụ: 180 giá trị lat
lon_var[:] = np.linspace(-180, 180, 360)  # Ví dụ: 360 giá trị lon

# Tạo dữ liệu nhiệt độ ngẫu nhiên để minh họa
data = np.random.rand(10, 180, 360)
data_var[:] = data

# Đóng file NetCDF
dataset.c
```
